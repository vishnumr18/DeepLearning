```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

```python
X, y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1,noise=80, random_state=13)
```

```python

plt.scatter(X, y);
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color='r')
```

```python
# Appling the gradient descent assuming the slopw value to be 78.35
# assuming a random value as b= 0
b=0
y_pred = ((78.35*X)+ b).reshape(4)
```

```python
# y_pred.reshape(4)
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='r', label='OLS')
plt.plot(X, y_pred, color='g', label='GD')
plt.legend()
plt.show()
```

```python
# Iteration 1
m = 78.35
b = 0

loss_slope = -2*np.sum(y - m*X.ravel()-b)
loss_slope
```

```python
# calculating the step size
lr = 0.1

step_size = lr*loss_slope
step_size
```

```python
# Update the value of b
b = b - step_size
b
```

```python
# Iteration 2
y_pred1 = ((78.35*X)+ b).reshape(4)
y_pred1
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='r', label='OLS')
plt.plot(X, y_pred1, color='#00a65a', label='b ={}'.format(b))
plt.plot(X, y_pred, color='#A3E4D7', label='b = 0')
plt.plot()
plt.legend()
plt.show()
```

```python
m = reg.coef_
b = 100
lr = 0.01

epoches = 100
for i in range(epoches):
    loss_slope = -2*np.sum(y- m*X.ravel()-b)
    step_size = lr*loss_slope
    b = b - step_size
    y_pred = m*X + b
    plt.plot(X, y_pred)

plt.scatter(X, y)
plt.show()
```
