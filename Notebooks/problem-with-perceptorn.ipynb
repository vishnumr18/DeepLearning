```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.linear_model import Perceptron
```

```python
and_data = pd.DataFrame()
or_data = pd.DataFrame()
xor_data = pd.DataFrame()
```

```python
and_data['input1'] = [1,1,0,0]
and_data['input2'] = [1,0,1,0]
and_data['output'] = [1,0,0,0]
```

```python
and_data
```

```python
or_data['input1'] = [1,1,0,0]
or_data['input2'] = [1,0,1,0]
or_data['output'] = [1,1,1,0]
```

```python
or_data
```

```python
xor_data['input1'] = [1,1,0,0]
xor_data['input2'] = [1,0,1,0]
xor_data['output'] = [0,1,1,0]
```

```python
xor_data
```

```python
sns.scatterplot(x=and_data['input1'], y=and_data['input2'], hue=and_data['output'],s=50);
```

```python
sns.scatterplot(x=or_data['input1'], y=or_data['input2'], hue=or_data['output'],s=50);
```

```python
sns.scatterplot(x=xor_data['input1'], y=xor_data['input2'], hue=xor_data['output'],s=50);
```

```python
# Instantiating the perceptron
clf1 = Perceptron()
clf2 = Perceptron()
clf3 = Perceptron()
```

```python
clf1.fit(and_data.iloc[:,0:2].values, and_data.iloc[:,-1].values)
```

```python

clf1.coef_, clf1.intercept_
```

```python
clf2.fit(or_data.iloc[:,0:2].values, or_data.iloc[:,-1].values)
```

```python

clf2.coef_, clf2.intercept_
```

```python
clf3.fit(xor_data.iloc[:,0:2].values, xor_data.iloc[:,-1].values)
```

```python

clf3.coef_, clf3.intercept_
```
